```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
plt.rcParams['font.sans-serif']='SimHei'
```

```python
iris_data_X=datasets.load_iris().data
iris_data_Y=datasets.load_iris().target.reshape(-1,1)
columns=datasets.load_iris().
label=datasets.load_iris().target_names
```

```python
data=pd.DataFrame(data=np.concatenate((iris_data_X,iris_data_Y),axis=1),index=None,
                  columns=['花萼长度','花萼宽度','花瓣长度','花瓣宽度','所属亚属'])
data.loc[data['所属亚属']==0,'所属亚属']='山鸢尾'
data.loc[data['所属亚属']==1,'所属亚属']='变色鸢尾'
data.loc[data['所属亚属']==2,'所属亚属']='维吉尼亚鸢尾'
data
```

```python
sns.pairplot(data, hue='所属亚属', markers='+',x_vars=["花萼长度", "花萼宽度"],y_vars=["花瓣长度", "花瓣宽度"])
plt.show()
```
